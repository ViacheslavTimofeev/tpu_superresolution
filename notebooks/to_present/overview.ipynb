# Постановка задачи

Разработка методов количественной оценки пористости и её пространственного распределения в керне является ключевой задачей цифровой обработки керна и цифровой петрофизики. Для традиционных коллекторов методы микротомографической визуализации (µCT) позволяют надёжно восстанавливать 3D-модель порового пространства и использовать её в последующих моделях фильтрации. Однако при работе с плотными коллекторами существенная доля пор имеет размеры меньше разрешения томографа. Такие субразрешаемые поры не фиксируются на стандартных изображениях и, как следствие, вносят значительные погрешности в расчётные оценки пористости и проницаемости.

Одним из подходов восстановления является задача создания изображений высокого разрешения (далее $HR$, high-resolution) из изображений низкого разрешения ($LR$, low-resolution), что в свою очередь делится на парный (supervised), непарный (unsupervised) и "само-надзорный" (self-supervised) процессы восстановления. Такие методики обычно называются задачей супер-разрешения (super-resolution, $SR$). Как было упомянуто, некоторая доля пор имеет меньшие размеры, чем разрешение томографа, поэтому для их обнаружения необходимо иметь дело с понятием trade-off между пространственным разрешением и размером образца. Данную проблему возможно решить, если использовать нейронные сети, обученные на заранее доступных $LR$ и $HR$. Это позволит увидеть субразрешенную пористость, корректно восстановить цифровую 3D копию керна большего размера, а также верно рассчитать долю пористости.

# Важные нюансы

1. Количество данных;
2. Являются ли LR и HR парными или нет;
3. Создаются ли LR из HR (синтетические) или являются настоящими;
4. Формат изображений: RGB или L;
