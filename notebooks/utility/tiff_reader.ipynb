```python
# 1) Импорт и выбор файла
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt

# укажи путь к файлу (например, один из сохранённых):
tiff_path_HR = Path("MRCCM2D/HR_test/block_ILS1_z0006_g0006.tiff")  # <-- поменяй при необходимости
tiff_path_LR = Path("MRCCM2D/LR_test/block_LR_z0006.tiff")  # <-- поменяй при необходимости
```

```python
# 2) Чтение через tifffile (предпочтительно для float32 TIFF)
import tifffile as tiff
img_HR = tiff.imread(tiff_path_HR)
img_LR = tiff.imread(tiff_path_LR)# np.ndarray

print(img_HR.shape, img_HR.dtype, float(np.nanmin(img_HR)), float(np.nanmax(img_HR)))
print(img_LR.shape, img_LR.dtype, float(np.nanmin(img_LR)), float(np.nanmax(img_LR)))
```

```python
plt.figure(figsize=(6,6))
plt.imshow(img_HR, cmap="gray")
plt.title(tiff_path_HR.name)
plt.axis("off")
plt.show()

plt.figure(figsize=(6,6))
plt.imshow(img_LR, cmap="gray")
plt.title(tiff_path_LR.name)
plt.axis("off")
plt.show()
```

```python
# 4) Гистограмма интенсивностей
vals_HR = img_HR[np.isfinite(img_HR)].ravel()
plt.figure(figsize=(6,3))
plt.hist(vals_HR, bins=256)
plt.xlabel("intensity")
plt.ylabel("count")
plt.show()

vals_LR = img_LR[np.isfinite(img_LR)].ravel()
plt.figure(figsize=(6,3))
plt.hist(vals_LR, bins=256)
plt.xlabel("intensity")
plt.ylabel("count")
plt.show()
```
