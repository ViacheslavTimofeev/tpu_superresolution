```python
from pathlib import Path
from typing import Dict

import numpy as np
from PIL import Image
import torch
from pytorch_msssim import ssim
import matplotlib.pyplot as plt

# устройство для вычислений
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

def load_gray_01(path: str | Path) -> torch.Tensor:
    """
    Загружает изображение как grayscale тензор [1,1,H,W] в диапазоне [0,1].
    """
    img = Image.open(path).convert("L")
    arr = np.asarray(img, dtype=np.float32) / 255.0  # [H,W]
    t = torch.from_numpy(arr).unsqueeze(0).unsqueeze(0)  # [1,1,H,W]
    return t.to(device)


def psnr(x: torch.Tensor, y: torch.Tensor, max_val: float = 1.0) -> float:
    """Ожидает тензоры в [0,1]. Возвращает средний PSNR по батчу."""
    mse = torch.mean((x - y) ** 2, dim=[1, 2, 3])  # [B]
    mse = torch.clamp(mse, min=1e-10)
    val = 20.0 * torch.log10(max_val / torch.sqrt(mse))  # [B]
    return float(val.mean())


def _show_triplet(hr: torch.Tensor, lr: torch.Tensor, pred: torch.Tensor, title: str):
    """
    Вспомогательная функция для отображения трёх изображений в один ряд.
    """
    imgs = [("LR", lr), ("HR", hr), ("PRED", pred)]

    plt.figure(figsize=(9, 3))
    for i, (label, tensor) in enumerate(imgs, start=1):
        img = tensor.squeeze().detach().cpu().numpy()  # [H,W]
        plt.subplot(1, 3, i)
        plt.imshow(img, cmap="gray", vmin=0.0, vmax=1.0)
        plt.title(label)
        plt.axis("off")

    plt.suptitle(title)
    plt.tight_layout()
    plt.show()


def eval_triplet(
    hr_path: str | Path,
    lr_path: str | Path,
    pred_path: str | Path,
    name: str | None = None,
) -> Dict[str, float]:
    """
    Считает PSNR и SSIM для:
    - LR vs HR  (bicubic baseline)
    - PRED vs HR (модель)
    и показывает все три изображения (LR/HR/PRED).

    Возвращает словарь с метриками.
    """
    hr = load_gray_01(hr_path)
    lr = load_gray_01(lr_path)
    pred = load_gray_01(pred_path)

    sample_name = name or Path(hr_path).stem

    with torch.no_grad():
        metrics = {
            "name": sample_name,
            "psnr_lr": psnr(lr, hr),
            "ssim_lr": float(ssim(lr, hr, data_range=1.0, size_average=True)),
            "psnr_pred": psnr(pred, hr),
            "ssim_pred": float(ssim(pred, hr, data_range=1.0, size_average=True)),
        }

    print(f"\n=== {metrics['name']} ===")
    print(f"LR   vs HR : PSNR = {metrics['psnr_lr']:.4f} dB, "
          f"SSIM = {metrics['ssim_lr']:.4f}")
    print(f"PRED vs HR : PSNR = {metrics['psnr_pred']:.4f} dB, "
          f"SSIM = {metrics['ssim_pred']:.4f}")

    # показать изображения
    _show_triplet(hr, lr, pred, title=sample_name)

    return metrics
```
